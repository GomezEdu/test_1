```python
import matplotlib.pyplot as plt
print("Setup terminado")
```

```python
marcas_clase = [1, 2, 3, 4, 5, 6]
frecuencias = [5, 15, 23, 37, 12, 8]

plt.figure(figsize = (12, 6))

plt.bar(marcas_clase, frecuencias,
       width = 1, edgecolor = "k", 
       color = ["#1974B4", "#465CBD", "#33A552", "#FFF80D", "#0DFFCF", "#D81919"])

plt.xlabel("Marcas de clase", fontsize = 20)
plt.ylabel("Frecuencias", fontsize = 20)
plt.title("Histograma", fontsize = 25)
plt.show()
```

```python
marcas_clase = [1, 2, 3, 4, 5, 6]
frecuencias = [5, 15, 23, 37, 12, 8]
marcas_texto = ["Cola", "Cola ligh", "Manzanita", "Fanta", "Jarrito de tutifruti", "Sprite"]
plt.figure(figsize = (12, 6))

plt.bar(marcas_clase, frecuencias,
       width = 1, edgecolor = "k", 
       color = ["#1974B4", "#465CBD", "#33A552", "#FFF80D", "#0DFFCF", "#D81919"])
plt.xticks(marcas_clase, marcas_texto, fontsize = 15, rotation = 45)
plt.xlabel("Marcas de clase", fontsize = 20)
plt.ylabel("Frecuencias", fontsize = 20)
plt.title("Histograma", fontsize = 25)
plt.grid()
plt.show()
```

```python
marcas_clase = [1, 2, 3, 4, 5, 6]
frecuencias = [5, 15, 23, 37, 12, 8]
marcas_texto = ["Cola", "Cola ligh", "Manzanita", "Fanta", "Jarrito de tutifruti", "Sprite"]
plt.figure(figsize = (12, 6))

plt.barh(marcas_clase, frecuencias,
       height = 0.75, edgecolor = "k", 
       color = ["#1974B4", "#465CBD", "#33A552", "#FFF80D", "#0DFFCF", "#D81919"])
plt.yticks(marcas_clase, marcas_texto, fontsize = 15, rotation = 45)
plt.xlabel("Marcas de clase", fontsize = 20)
plt.ylabel("Frecuencias", fontsize = 20)
plt.title("Diagrama de barras", fontsize = 25)
plt.grid()
plt.show()
```

```python
marcas_clase = [1, 2, 3, 4, 5, 6] # Datos del eje x
frecuencias = [0.05, 0.15, 0.23, 0.37, 0.12, 0.08] # Datos del eje y
frecuencias = [0.05, 0.2, 0.43, 0.8, 0.92, 1] # Datos del eje y
marcas_texto = ["Cola", "Cola ligh", "Manzanita", 
                "Fanta", "Jarrito de tutifruti", "Sprite"] # Etiquetas para el eje x

# Ajustes necesarios para trazar el poligono "NO TOCAR"
datos_x = [0] + marcas_clase 
datos_y = [0] + frecuencias 

plt.figure(figsize = (12, 6)) # Ancho, alto de la figura

# Este Ojiva cuenta con: 
# - Linea punteada (linestyle = ":"). https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
# - Marcador triangular ( marker = "v"). https://matplotlib.org/stable/api/markers_api.html

plt.plot(datos_x, datos_y,  
         linewidth = 5, color = "r", linestyle = ":",
        marker = "v", markersize = 15, markerfacecolor = "y", markeredgecolor = "b")
         
plt.xticks(marcas_clase, marcas_texto, fontsize = 15, rotation = 45) # Colocar texto sobre el eje x
plt.xlabel("Marcas de clase", fontsize = 20) # Etiqueta del eje x
plt.ylabel("Frecuencias", fontsize = 20) # Etiqueta del eje y
plt.title("Ojiva", fontsize = 25) # Titulo del grafico
plt.grid() # Se activa la cuadricula sobre el grafico
plt.show() # Se muestra el grafico
```

```python
datos = [8 , 12, 25, 15, 23, 17]
marcas_texto = ["Cola", "Cola ligh", "Manzanita", "Fanta", "Jarrito de tutifruti", "Sprite"]
separaciones = [0, 0, 0.2, 0, 0, 0]

plt.pie(datos, 
       explode = separaciones, 
       counterclock = False, 
       startangle = 90, 
       autopct = "%0.1f%%", 
       pctdistance = 0.8, 
       colors = ["#1974B4", "#465CBD", "#33A552", "#FFF80D", "#0DFFCF", "#D81919"], 
       labels = marcas_texto)
plt.show()
```
